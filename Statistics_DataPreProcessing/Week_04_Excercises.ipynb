### Grouping Data by Categorical Values

**were delays shorter or longer on some airlines as opposed to others?**
To compare delays across airlines, we need to group the records of airlines together. Boolean indexing won’t work for this—it can only separate the data into two categories: one that is true, and one that is false (or, in this case, one that is delayed and one that is not delayed). What we need here is two categories (delayed and not delayed) for each airline.

**GroupBy** splits the data into multiple buckets based on the input

**.groupby()**

```python
#Calculate Group By
group_by_carrier=df.groupby(['UNIQUE_CARRIER','DELAY'])
```

-  groupby() splits the dataset data into buckets by carrier (‘unique_carrier’)
-  It then
-  splitting the records inside each carrier bucket into delayed or not delayed (‘delayed’). The result is assigned to the group_by_carrier variable.

**Check DataType of group_by_carrier Object**

```python
print(group_by_carrier)
```

- It is an Object!
-  We can think of that as instructions on how to group, but without instructions on how to display values
-  We need to provide instructions on what values to display. For example:
-  We’re grouping all of the rows that share the same carrier, as well as all the rows that share the same value for delayed. We need to tell the function what to do with the other values. We could do any number of things:

    -  average the number of minutes of delay
    -  sum the total number of minutes of delay
-   Instead of averaging or summing, use .size() to count the number of rows in each grouping:

```python
#Apply Size Method on this object
group_by_carrier.size()
```

-  Pandas has a handy **.unstack()** method
-  We can use it to convert the results into a more readable format and store that as a new variable

```python
#Unstack Method for Readable format represention of Group By Datya
count_delays_by_carrier=group_by_carrier.size().unstack()
count_delays_by_carrier
```

### Plotting Grouped Data

```python
count_delays_by_carrier.plot(kind='barh',stacked=True,colormap='winter', figsize=[16,6] )
```

## Pivot Tables

A pivot table is composed of counts, sums, or other aggregations derived from a table of data. You may have used this feature in spreadsheets, where you would choose the rows and columns to aggregate on, and the values for those rows and columns. It allows us to summarize data as grouped by different values, including values in categorical columns.

```python
#Calculate Total No of Flights/carrier
flights_by_carrier=df.pivot_table(index='FL_DATE',columns='UNIQUE_CARRIER',values='FL_NUM',aggfunc='count')
flights_by_carrier.head(10)
```

```python
#Calculate Average of Numerical Values using Pivot Table
df.pivot_table(columns= 'FL_DATE',aggfunc='sum')
```

## numpy .arange()

Say you’re interested in analyzing length of delays and you want to put these lengths into bins that represent every 10 minute period. You can use the numpy method .arange() to create a list of numbers that define those bins. The bins of ten minute intervals will range from 50 minutes early (-50) to 200 minutes late (200). The first bin will hold a count of flights that arrived between 50 and 40 minutes early, then 40 and 30 minutes, and so on

```python
bin_values = np.arange(start=-50, stop=200, step=10)
```

- Let’s plot the distribution of Southwest flight delays as a histogram using the bin_values variable created above:

```python
wn_carrier = df[df['UNIQUE_CARRIER'] == 'WN']
wn_carrier['ARR_DELAY'].hist(bins=bin_values, figsize=[14,6])
```

Great! We can see that the vast majority of Southwest flights weren’t more than 30 minutes late. Though the airline has many flights, the majority of them aren’t late enough.

## Comparing distributions with histograms

-  Seeing one distribution is helpful to give us a shape of the data, but how about two?

-  **Compare the distribution of two airlines with a similar number of total flights, US Airways and Envoy Air:**

```python
bin_values = np.arange(start=-50, stop=200, step=10)
us_mq_airlines = df[df['UNIQUE_CARRIER'].isin(['UA','DL'])] # create index of flights from those airlines

group_carriers = us_mq_airlines.groupby('UNIQUE_CARRIER')['ARR_DELAY'] # group values by carrier, select minutes delayed

group_carriers.plot(kind='hist', bins=bin_values, figsize=[12,6], alpha=.4, legend=True) # alpha for transparency
```

**Key Findings**
The two distributions look similar, but not the same 
-  Let's use Describe method to didcern the different between these two Carrier datasets

```python
group_carriers.describe()
```

Above result Indicates that **delays on United Airlines are more distributed than delays on Delta Airways**, meaning that the values are more spread out.
